```python
import pandas as pd
import re

df=pd.DataFrame([])

def extraire_nombres(texte):
    # Utilisation d'une expression régulière pour trouver tous les nombres dans le texte
    nombres = re.findall(r'\d{1,3}(?:\s?\d{3})*', texte)
    # Convertir les nombres de chaînes de caractères en entiers
    nombres = [int(nombre.replace('\u202f', '')) for nombre in nombres]  # Enlever l'espace insécable (\u202f) des milliers
    return nombres
```
